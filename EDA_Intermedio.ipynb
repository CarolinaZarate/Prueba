```python
iris = load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names) #transformación de datos
df_iris['target'] = iris.target #target es la columna que se quiere predecir
print(df_iris)
```

```python
df_iris.rename({'target':'species'}, inplace=True, axis=1)
print(df_iris)
```

```python
df_iris['species'] = df_iris['species'].astype('category')
df_iris.info()
```

```python
#Asigno valores nulos a la columna de especies
iris_amp = mf.ampute_data(df_iris, perc=0.25, random_state=100)
iris_amp
```

```python
#Crear Kernel
kds = mf.ImputationKernel(
    iris_amp,
    random_state=100 #profundidad del árbol
)

kds.mice(2) #correr 2 iteraciones

iris_completo = kds.complete_data()

iris_completo
```
